# Simple block representing a simplified car motion model

Simulink block model represents a mathematical relationship between its input and output. The simulink block model shown below represents integration of a brief pulse twice to get a ramp. The results display in a Scope window. The input pulse represents a press of the gas pedal â€” 1 when the pedal is pressed and 0 when it is not. The output ramp is the increasing distance from the starting point.

Let's us consider an obstacle lies at a distance of 20 unit ahead from the car. The model of the car must be refined to make a hard stop close to 20 unit before hitting the obstacle. The block model below shows a refined model with integrator set with upper limit at x = 20 and a proximity sensor model used to achieve the braking task. 

The block model shows an updated model to solve the problem of obstacle involved.The plot above shows how initially the obstacle is at 20 m and starts getting closer with time. To make the simulation close to a real test run compared to the simulated version, a band-limited white noise is added to the sensor and the digital sensor is set to 0.1 seconds of firing rate.
